```python
import pandas as pd
from matplotlib.font_manager import FontProperties

# Read the CSV file
df = pd.read_csv('free_energy.csv',header=None)
df1 = pd.read_csv('free_energy_10nm_phi0.01_DJA.csv',header=None)
df2 = pd.read_csv('free_energy_10nm_phi0.01_retarded.csv',header=None)
# Rename the columns
df.rename(columns={0: 'Time', 1: 'Free_Energy'}, inplace=True)
df1.rename(columns={0: 'Time', 1: 'Free_Energy'}, inplace=True)
df2.rename(columns={0: 'Time', 1: 'Free_Energy'}, inplace=True)

# # Plot the data
# plt.plot(df['Time'], df['Free_Energy'],  linewidth=2, label='Fully ret')
# plt.plot(df1['Time'], df1['Free_Energy'], linewidth=2, label=r'$\mu_{\mathrm{mix}} = 1$')
# plt.plot(df2['Time'], df2['Free_Energy'], linewidth=2, label=r'$c \rightarrow \infty$')

# plt.axhline(y=0, color='k', linestyle='--')
# plt.xlabel('Dist (nm)',fontsize=17)
# plt.ylabel(r'$\mathrm{Force/Radius}\, (\mu\mathrm{N}/\mathrm{m})$', fontsize=17)
# plt.yscale('symlog', linthresh=10**(-9))

# # plt.title('Free Energy vs Dist')
# plt.yticks([-1e-2,-1e-5, -1e-8,  0, 1e-8, 1e-5, 1e-2],['$-10^{-2}$','$-10^{-5}$', '$-10^{-8}$',  '0', '$10^{-8}$', '$10^{-5}$', '$10^{-2}$'])
# plt.gca().spines['top'].set_linewidth(2)
# plt.gca().spines['right'].set_linewidth(2)
# plt.gca().spines['bottom'].set_linewidth(2)
# plt.gca().spines['left'].set_linewidth(2)
# plt.ylim(-1e-5, 1e-1)
# plt.xlim(-2, 1500)
# plt.xticks(fontsize=14)
# plt.yticks(fontsize=14)
# plt.minorticks_on()

# Use a more beautiful style
# plt.style.use('seaborn-dark')
# plt.style.use('ggplot')

# Increase figure size
plt.figure(figsize=(8, 6))
# plt.rcParams['font.weight'] = 'bold'
# plt.rcParams['axes.labelweight'] = 'bold'
# plt.rcParams['mathtext.default'] = 'regular'  # Ensures consistent font rendering
plt.rcParams['mathtext.bf'] = 'bold' 
# plt.gca().minorticks_on()
# plt.gca().tick_params(axis='both', which='minor', length=4, width=1)


# Plot the data with updated style
plt.plot(df['Time'], df['Free_Energy'], linewidth=2, label='Fully ret')
plt.plot(df1['Time'], df1['Free_Energy'], linewidth=2, label=r'$\mu_{\mathrm{mix}} = 1$')
plt.plot(df2['Time'], df2['Free_Energy'], linewidth=2, label=r'$c \rightarrow \infty$')

plt.axhline(y=0, color='k', linestyle='--')
plt.xlabel('Dist (nm)', fontsize=17)
plt.ylabel(r'$\mathrm{Force/Radius}\, (\mu\mathrm{N}/\mathrm{m})$', fontsize=17)
plt.yscale('symlog', linthresh=10**(-9))


plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)




plt.yticks([-1e-2, -1e-5, -1e-8, 0, 1e-8, 1e-5, 1e-2], [r'$\mathbf{-10^{-2}}$', r'$-10^{-5}$', r'$-10^{-8}$', r'0', r'$10^{-8}$', r'$10^{-5}$', r'$10^{-2}$'])
plt.ylim(-1e-5, 1e-1)
plt.xlim(-2, 1500)
plt.title(r'$\phi =0.01 \quad \mathrm{and} \quad \mathrm{Radius} = 10 \,\mathrm{nm}$', fontsize=17, fontweight='bold')
# plt.xticks(fontsize=12, fontweight='bold')
# plt.yticks(fontsize=12, fontweight='bold')
plt.tick_params(axis='both', which='major', length=6, width=2, labelsize=14)
plt.tick_params(axis='both', which='minor', length=4, width=1)

plt.minorticks_on()
plt.legend(fontsize=14, loc='upper center',  ncol=3)
# plt.savefig('free_energy.png',bbox_inches='tight', dpi=2000)
plt.show()
```

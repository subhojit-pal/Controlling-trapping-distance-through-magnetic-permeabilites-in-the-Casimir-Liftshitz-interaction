```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams["mathtext.fontset"] = "cm"
plt.rcParams['mathtext.bf'] = 'bold' 
# plt.rcParams['font.weight'] = 'bold'
plt.rcParams['mathtext.default'] = 'regular'
# plt.rcParams['axes.labelweight'] = 'bold'
# plt.rcParams.update({
#     "text.usetex": True,
#     "font.family": "serif"
# })
# plt.rc('text', usetex=True)
df3 = pd.read_csv('free_energy_5nm_phi0.01.csv',header=None)
df = pd.read_csv('free_energy.csv',header=None)
df1 = pd.read_csv('free_energy_15nm_phi0.01.csv',header=None)
df2 = pd.read_csv('free_energy_20nm_phi0.01.csv',header=None)

df.rename(columns={0: 'Time', 1: 'Free_Energy'}, inplace=True)
df1.rename(columns={0: 'Time', 1: 'Free_Energy'}, inplace=True)
df2.rename(columns={0: 'Time', 1: 'Free_Energy'}, inplace=True)
df3.rename(columns={0: 'Time', 1: 'Free_Energy'}, inplace=True)

plt.figure(figsize=(8, 6))
plt.gca().spines['top'].set_linewidth(2)
plt.gca().spines['right'].set_linewidth(2)
plt.gca().spines['bottom'].set_linewidth(2)
plt.gca().spines['left'].set_linewidth(2)

# x = np.linspace(0, 10, 100)
# y = np.sin(x)
plt.plot(df3['Time'], df3['Free_Energy'], linewidth=2, label=r'$5\,\mathrm{nm}$')
plt.plot(df['Time'], df['Free_Energy'], linewidth=2, label=r'$10 \,\mathrm{nm}$')
plt.plot(df1['Time'], df1['Free_Energy'], linewidth=2, label=r'$15\,\mathrm{nm}$')
plt.plot(df2['Time'], df2['Free_Energy'], linewidth=2, label=r'$20\,\mathrm{nm}$')

plt.minorticks_on()

# plt.plot(x, y, label=r'$\mathregular{\epsilon}$')  # Use \epsilon for the straight version
plt.legend(fontsize=18)
# plt.yticks(fontsize=12, fontweight='bold')
# plt.xticks(fontsize=12, fontweight='bold')
plt.xlabel(r'$\mathrm{\mathregular{\xi}}$', fontsize=20)  # Example in axis labels
plt.ylabel(r'$\mathregular{\epsilon}(\mathbf{\mathrm{i}} \mathrm{\xi}_{m})$', fontsize=20)  # Example in axis labels
plt.tick_params(axis='both', which='major', length=6, width=2, labelsize=14)
plt.tick_params(axis='both', which='minor', length=4, width=1.15)
# plt.ylabel(r'$U+3B5$', fontsize=14)
plt.title(r'Straight Epsilon: $\epsilon$', fontsize=16)
plt.show()
```
